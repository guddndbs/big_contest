```python
import pandas as pd
loan_result = pd.read_csv("loan_result.csv")
log_data = pd.read_csv("log_data.csv")
user_spec = pd.read_csv("user_spec.csv",encoding='CP949')
```

```python
not_nan = loan_result[loan_result['is_applied'].notnull()]
```

```python
sorted_nan = not_nan.sort_values(by='application_id')

sorted_spec = user_spec.sort_values(by='application_id')
```

```python
sorted_nan.head()
```

```python
sorted_spec.head()
```

```python
merge_df = pd.merge(left=sorted_nan, right=sorted_spec, how="inner", on="application_id")
```

```python
merge_df.head()
```

```python
merge_arr = merge_df.to_numpy()
print(merge_arr)
```

```python
def date_to_num(date1,date2):
    year1 = date1[:4]
    month1 = date1[4:6]
    year2 = date2[:4]
    month2 = date2[5:7]
    return (int(year2) - int(year1)) * 12 + (int(month2) - int(month1))
```

```python
for i in range(len(merge_arr)):
    if merge_arr[i][14] >= 100000:
        merge_arr[i][14] = date_to_num(str(merge_arr[i][14]),merge_arr[i][1])
```
